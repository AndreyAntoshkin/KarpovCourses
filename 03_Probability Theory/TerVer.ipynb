```python
from math import factorial

def binominalniy_kef (n, k):
    return factorial(n) / (factorial(k) * factorial(n-k))


binominalniy_kef(10, 3)
```

- В магазине 7 покупателей. Каждый может совершить покупку с вероятностью 0.4.

**Найдите вероятность того, что не более двух человек совершат покупку.**

Ответ округлите до 2 знака.

```python
n = 7 
p = 0.4
q = 1 - p
sum_ver = 0
for i in range(0, 3):
    sum_ver += binominalniy_kef(n, i) * p**i * q**(n-i)
print(f'Вероятность того, что не более двух человек совершат покупку: {round(sum_ver,2)}')
```

- Для нормальной работы отдела по дата-аналитике необходимо, чтобы в офисе было не менее восьми программистов, а всего в штате 10 программистов.

**Вероятность того, что программист заболеет, равна 0.1. Найдите вероятность нормальной работы отдела по дата-аналитике.**

Ответ округлите до 2 знака.

```python
p = 0.1
q = 1 - p
n = 10
k = [8, 9, 10]
sum_ver = 0

for i in k:
    sum_ver += binominalniy_kef(n, i) * q**i * p**(n-i)
print(f'Вероятность нормальной работы отдела по дата-аналитике: {round(sum_ver,2)}')
```

- Вероятность того, что день рождения наугад взятого участника чата приходится на определенный день года,
- равна 1/365 для каждого из 365 дней. 
- Найдите наиболее вероятное число счастливчиков, которые родились 1 марта.

Для решения этой задачи вам может потребоваться эта формула.

np−q≤k≤np+p,q=1−p.

```python
n = 1730 
p = 1/365
q = 1 - p

left_part = n * p - q
right_part = n * p + p
print(f'{left_part} <= k <= {right_part}')
```
