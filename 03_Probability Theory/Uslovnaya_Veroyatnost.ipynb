Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.

Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что на радаре нет самолета?

Ответ округлите до 4 знаков.

Обратите внимание: в отличие от задачи про медицинский тест, здесь фигурирует два различных типа ошибок:
Ложное обнаружение самолёта (5%).
Пропуск самолёта (2%).
В остальном логика такая же, как и в задаче про медицинский тест. В данном случае нас интересуют два события:
Самолёта нет И не было ложного срабатывания.
Самолёт есть И его пропустили (должно быть в числителе).

```python
obnar = 0.98 # самолет на радаре и обнаруден с вероятностью

osechka = 0.02 # самолет на радаре но при это не обнаружен с вероятностью

osechka_2 = 0.05 # самолета нет на радаре но при этом он обнаружен

nahozhdenie = 0.07 # самолет находистя в зоне радара вероятность 

#Зона действия радара = 5% осечка + 7% * 0.98 

n_1 = 0.07 # самолет находится в зоне действия радара
n_2 = 0.93 # самолет вне действия радара

P_A = n_1 * obnar + n_2 * osechka # Самолет находится в зоне действия радара + осечка что самолет не находится

P = (nahozhdenie * osechka) / ((nahozhdenie * osechka) + (1-nahozhdenie * 1-osechka_2))

round(P,4)
```

Система РЛС (радиолокационная система) может обнаружить присутствие самолета в 98% случаев. Однако если в зоне действия радара нет самолета, система все равно сообщает (ложно), что самолет присутствует, в 5% случаев. В любой момент времени вероятность того, что самолет находится в зоне действия радара, составляет 7%.

Какова вероятность того, что самолет находится в зоне действия радара, учитывая, что радар показывает наличие самолета?

Ответ округлите до 3 знаков.

```python
obnar = 0.98 # самолет на радаре и обнаруден с вероятностью

osechka = 0.02 # самолет на радаре но при это не обнаружен с вероятностью

osechka_2 = 0.05 # самолета нет на радаре но при этом он обнаружен

nahozhdenie = 0.07 # самолет находистя в зоне радара вероятность 


P_A = (nahozhdenie * obnar) + ((1 - nahozhdenie) * osechka_2)

P = (nahozhdenie * obnar) / P_A

round(P,3)
```
