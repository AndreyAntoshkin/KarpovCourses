```python
import seaborn as sns

tips = sns.load_dataset("tips")
tips
```

```python
bill_mean = tips['total_bill'].mean()
bill_mean
```

```python
bill_std = tips['total_bill'].std()
bill_std
```

```python
tips['z-values_bill'] = (tips['total_bill'] - bill_mean) / bill_std
```

```python
tips
```

```python
tips["z-values_bill_scipy"] = st.zscore(tips["total_bill"], ddof=1)
```

```python
from scipy.stats import norm

# Для стандартного нормального распределения (μ=0, σ=1)
print(norm.cdf(0)) # Выведет 0.5 (вероятность быть меньше 0)

# Для произвольного распределения
mean = 100
std_dev = 15
print(norm.cdf(115, loc=mean, scale=std_dev)) # Вероятность быть меньше 115
```

Вы провели расчеты и выяснили, что среди студентов курса по статистике распределение роста является нормальным со средним значением, равным 173, и стандартным отклонением, равным 11.

Приблизительно какой процент студентов имеет рост на интервале от 181 до 192? Ответ округлите до целого.

```python
import scipy.stats as st
less_than_181 = st.norm.cdf(loc=173, scale=11, x=181)
less_than_192 = st.norm.cdf(loc=173, scale=11, x=192)

less_than_192 - less_than_181
```
